Теперь надо нарисовать автокорреляционные функции и провести тесты Дики-Фуллера на стационарность рядов, а также посмотреть на кросс-корреляцию cap rates и регрессоров.

```python
import statsmodels.api as sm

def tsdisplay(y:pd.Series, 
              lags:int, 
              series_name:str,
              figsize:tuple=(14,5), 
              ):
    """
    Plots series supplied, its sample ACF and PACF 
    """
    fig = plt.figure(figsize=(14,5))
    y.plot(ax = fig.add_subplot(211), title = series_name)
    fig = sm.graphics.tsa.plot_acf(y, lags = lags, ax = fig.add_subplot(223))
    fig = sm.graphics.tsa.plot_pacf(y, lags = lags, ax = fig.add_subplot(224))
    plt.tight_layout()
```

```python
titles=dict(zip(d.columns, 
                ['Office cap rate', 'Retail cap rate',
                 'Logistics cap rate', 'OFZ 5 year YTM',
                 'CBR key rate', 'Russia CPI y-o-y']))

for series in d.columns:
    tsdisplay(d.retail, 
              lags=10, 
              figsize=(14,5), 
              series_name=titles[series])
```

Можно сделать интересные наблюдения: 
* ставки по недвижимости гораздо меее волатильные, чем предикторы, и явная зависимость между ними и ОФЗ пропадает при высоких доходностях облигаций (10%+)
* они явно реагируют на изменения регрессоров с определенным лагом
* выраженной сезонности в ставках для недвижимости нет
* три cap rates явно коинтегрированы между собой, причем логистическая недвижимость всегда предлагает премию к офисной и розничной недвижимости, а розничная - к офисной

Также было бы интересно смоделировать сразу и доходность ОФЗ, которая может зависеть от CDS, поэтому выгрузим данные по CDS с инвестинга. Хотя не будем торопиться с этим регрессором, ведь сейчас, по идее, ОФЗ и так можно смоделировать через сумму константы, ставки и инфляции.

Итак, данные у нас квартальные, всего около 60 наблюдений. Как было бы разумно сформулировать модель? Пусть $O$ -  cap rate для офисов, $R$ - розничная cap rate, $L$ - для логистики, $\mu$ - некоторая константа. Тогда

$$
\begin{cases}
\bigtriangleup O_t = \alpha( -O_t + \mu + \beta_{1} R_t + \beta_{2} L_t + \beta_{3} OFZ_t) + \beta_6 \bigtriangleup O_{t-1} + \varepsilon_t \\[5pt]
\bigtriangleup R_t = \alpha( -O_t + \mu + \beta_{1} R_t + \beta_{2} L_t + \beta_{3} OFZ_t) + \beta_7 \bigtriangleup R_{t-1} + \varepsilon_t \\[5pt]
\bigtriangleup L_t = \alpha( -O_t + \mu + \beta_{1} R_t + \beta_{2} L_t + \beta_{3} OFZ_t) + \beta_8 \bigtriangleup L_{t-1} + \varepsilon_t \\[5pt]
\bigtriangleup OFZ_t = \alpha( -OFZ_t + \mu + \beta_{4} CPI_{t-1} + \beta_{5} CBR_t) + \beta_9 \bigtriangleup OFZ_{t-1} + \varepsilon_t \\
\end{cases}
$$

То есть мы предполагаем, что:
1. Ставки на рынке недвижимости связаны между собой, ОФЗ и ИПЦ коинтеграционным соотношением, а также зависят от собственных первых лагов (значение квартал назад)
2. Доходность ОФЗ определяется только ключевой ставкой, неким спредом к ней и ИПЦ - это, конечно, не так, особенно на коротком горизонте, но поскольку данных мало, коэффициентов много, то нет желания усложнять, добавляя еще и term premium.
3. Таким образом, с помощью двух экзогенных регрессоров - ключевой ставки и ИПЦ - мы сможем смоделировать YMT ОФЗ, а с ее помощью и с помощью ИПЦ - долгосрочную динамику cap rates для всех классов недвижимости.
